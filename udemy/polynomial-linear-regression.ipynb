```python
# Import modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Importing dataset
dataset = pd.read_csv('datasets/polynomial-linear-regression-data.csv')

# Separate dataset (X) and dependent vector (y)
X = dataset.drop(columns=['Position', 'Salary'])
y = dataset.filter(items=['Salary'])

# Small amount of data, so we want to include all the data
```

```python
# Fitting Linear regression to the dataset for reference to new polynomial regression model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)


# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
pr = PolynomialFeatures(degree=3)
X_poly = pd.DataFrame(pr.fit_transform(X), columns=['Level^0', 'Level^1', 'Level^2', 'Level^3'])
lr_2 = LinearRegression()
lr_2.fit(X_poly, y)

# Visualize result of simple linear regression
plt.scatter(X, y, color='red')
plt.plot(X, lr.predict(X), color='blue')
plt.title('Salary vs. Level (Simple Linear Regression)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()

# Visualize result of polynomial linear regression
X_interp = np.arange(X['Level'].min(), X['Level'].max(), 0.01)
X_interp = X_interp.reshape(-1, 1)
plt.scatter(X, y, color='red')
plt.plot(X_interp, lr_2.predict(pr.fit_transform(X_interp)), color='blue')
plt.title('Salary vs. Level (Polynomial Linear Regression)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()


```
