## Data Pre-Processing

```python
# Import modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Importing dataset
dataset = pd.read_csv('datasets/decision-tree-regression-data.csv')

# Separate dataset (X) and dependent vector (y)
X = dataset.drop(columns=['Position', 'Salary'])
y = dataset.filter(items=['Salary'])

# Small amount of data, so we want to include all the data
# Decision Tree Regression library does feature scaling for us, 
# so there is no need to manually feature scale
```

## Decision Forest Regression

```python
# Fitting random forest regression to the dataset
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=300, random_state=0)
rfr.fit(X, y)

# Predicting a new result
y_pred = rfr.predict([[6.5]])
print(y_pred)

# Visualize result of svr regression
X_interp = np.arange(X['Level'].min(), X['Level'].max(), 0.01)
X_interp = X_interp.reshape(-1, 1)
plt.scatter(X, y, color='red')
plt.plot(X_interp, rfr.predict(X_interp), color='blue')
plt.title('Salary vs. Level (Random Forest Regression)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```
