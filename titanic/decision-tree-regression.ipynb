## Data Pre-Processing

```python
# Import modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Importing dataset
dataset = pd.read_csv('datasets/full.csv')

# Separate dataset (X) and dependent vector (y)
X = dataset.filter(items=['Pclass'])
y = dataset.filter(items=['Survived'])

# Handle Missing values
## Filling missing
X.fillna(X.mean(), inplace=True)

# Splitting Training and Test Set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Decision Tree Regression

```python
# Fitting SVR to dataset
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)

# Visualize result of support vector regression
X_test_interp = np.arange(X_test['Pclass'].min(), X_test['Pclass'].max(), 0.01)
X_test_interp = X_test_interp.reshape(-1, 1)
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test_interp, dtr.predict(X_test_interp), color='blue')
plt.title('Survival vs. Pclass (Decision Tree Regression)')
plt.xlabel('Pclass')
plt.ylabel('Survival')
plt.show()
```
